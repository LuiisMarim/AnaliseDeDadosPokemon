```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn import datasets
data = pd.read_csv('pokemon.csv', header=(0))
data.drop(['is_mega','is_legendary', 'is_mythical', 'base_happiness', 'base_egg_steps','capture_rate','against_ghost',	'against_grass','against_ground','against_ice','against_normal','against_poison','against_psychic','against_rock','against_steel','against_water','against_bug','against_dark','against_dragon','against_electric','against_fairy','against_fighting','against_fire','against_flying','generation','classfication','abilities','height_m','weight_kg','type2'	],axis = 1, inplace = True)
print("Número de linhas e colunas:",data.shape)
data.head(25)
```

```python
data = data.to_numpy()
y = data[:,3]
print("X = Hp")
print("Y = Base_total")
x = data[:,4]
X = np.array(x).reshape(-1, 1)
Y = np.array(y)
modelo = LinearRegression().fit(X, Y)
print("Interceptação: ", modelo.intercept_)
print ("Coeficiente de Correlação: ", modelo.coef_)
plt.scatter(X, Y)
plt.plot(X, modelo.predict(X), color = 'red')
```

```python
from sklearn.metrics import r2_score
y_pred=modelo.predict(X)
R2 = r2_score(y, y_pred)
print('R2:', R2)
```

```python
n = np.size(x) 
Sx= np.sum(x)
Sy=np.sum(y)
Sxy=np.sum(x*y)
Sxx=np.sum(x*x)   
a1=(n*Sxy-Sx*Sy)/(n*Sxx-Sx**2)   #Calcula o coeficiente a1 da reta
a0=(Sxx*Sy-Sxy*Sx)/(n*Sxx-Sx**2) #Calcula o coeficiente a0 da reta
print(a1)
print(a0)
```

```python
# mostra os dados
plt.scatter(x, y, color = "b", marker = "o", s = 50) 
  # prediz os valores
y_pred = a0 + a1*x 
    # mostra a reta de regressão
plt.plot(x, y_pred, color = "r") 
  
plt.xlabel('x', fontsize = 15) 
plt.ylabel('y', fontsize = 15) 
plt.show(True) 
```

```python
#REGRESSÃO LINEAR MÚLTIPLA

y = data[:,3]
x = data[:,4:9]
#X = np.array(x).reshape(-1, 1)
y = np.array(y)
```

```python
from sklearn.linear_model import LinearRegression

# modelo de regressão linear múltipla
modelo = LinearRegression().fit(x, y)

y_pred = modelo.predict(x)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
l = plt.plot(y_pred, y, 'bo')
plt.setp(l, markersize=10)
plt.setp(l, markerfacecolor='C0')
plt.ylabel("y", fontsize=15)
plt.xlabel("Prediction", fontsize=15)

# mostra os valores preditos e originais
xl = np.arange(min(y), 1.2*max(y),(max(y)-min(y))/10)

yl = xl
plt.plot(xl, yl, 'r--')
plt.show(True)
```

```python
from sklearn.metrics import r2_score
R2 = r2_score(y, y_pred)
print('R2:', R2)
```
